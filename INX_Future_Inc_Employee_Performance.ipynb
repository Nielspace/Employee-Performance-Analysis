# Age Grouping

- Grouping helps to reduce the numbers and put into set for easy and better understanding of data. 

```python
for dataset in data:
    dataset['Age'] = dataset['Age'].astype(int)
    dataset.loc[ dataset['Age'] <= 22, 'Age'] = 0 
    dataset.loc[(dataset['Age'] > 22) & (dataset['Age'] <= 26), 'Age'] = 1 
    
    dataset.loc[(dataset['Age'] > 26) & (dataset['Age'] <= 30), 'Age'] = 2 
    dataset.loc[(dataset['Age'] > 30) & (dataset['Age'] <= 35), 'Age'] = 3
    dataset.loc[(dataset['Age'] > 35) & (dataset['Age'] <= 40), 'Age'] = 4
    dataset.loc[(dataset['Age'] > 40) & (dataset['Age'] <= 45), 'Age'] = 5
    dataset.loc[(dataset['Age'] > 45) & (dataset['Age'] <= 55), 'Age'] = 6
    dataset.loc[ dataset['Age'] > 55, 'Age'] = 7
```

```python
Age=df['Age']
sb.distplot(Age)
```

- Most of the graduates are above 22
- New comers tends to work for only 4 years or less before changing the job
- People who get good performance rating are usually between 30 to 40 years of age. This group of people are hard working 
- People clossing to retirement doesnot get good ratings because maybe they have achieved so much that they tend loosen up a bit and enjoy the easier and successful life.

# Distance Grouping



```python
for dataset in data:
    dataset['DistanceFromHome']=dataset['DistanceFromHome'].astype(int)
    dataset.loc[ dataset['DistanceFromHome'] <= 5, 'DistanceFromHome'] = 0
    dataset.loc[(dataset['DistanceFromHome'] > 5) & (dataset['DistanceFromHome'] <= 10), 'DistanceFromHome'] = 1
    dataset.loc[(dataset['DistanceFromHome'] > 10) & (dataset['DistanceFromHome'] <= 15), 'DistanceFromHome'] = 2
    dataset.loc[(dataset['DistanceFromHome'] > 15) & (dataset['DistanceFromHome'] <= 20), 'DistanceFromHome'] = 3
    dataset.loc[(dataset['DistanceFromHome'] > 20) & (dataset['DistanceFromHome'] <= 25), 'DistanceFromHome'] = 4
    
    dataset.loc[ dataset['DistanceFromHome'] > 25, 'DistanceFromHome'] = 5
```

```python
from sklearn.model_selection import train_test_split, cross_val_score, KFold
```

To reduce and avoid overfitting I am using <b>cross_val_score </b> and <b> KFold </b> to use all the split train/test data for evaluation

## K_nearest Neighbours

The reason I am using nearest neighbours because it determines how close our sample is to the training set. 

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
kfold=KFold(n_splits=20, shuffle=True, random_state=23)
```

```python
clf=KNeighborsClassifier(n_neighbors=5)
```

```python
score = cross_val_score(clf, train, target, cv=kfold, n_jobs=1)
```

```python
score
```

```python
round(np.mean(score)*100,2)
```

```python
clf.fit(train,target)
```

```python
clf.score(train,target)
```

## Random_forrest C

<p>
    I Used random forrest because it is quite versatile and it is an enchanced version of decision tree. In Decision Tree a principle node is taken and decisions are made based on yes or no. In Random Forrest the we can have multiple decision trees which avoids overfitting. 
    
   Random forests creates decision trees on randomly selected data samples, gets prediction from each tree and selects the best solution by means of <b> voting </b>. It also provides a pretty good indicator of the feature importance.
   
   It does not suffer from the overfitting problem. The main reason is that it takes the average of all the predictions, which cancels out the biases.
    

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
clf4=RandomForestClassifier(n_estimators=26, criterion='entropy', max_depth=None, min_samples_split=12, min_samples_leaf=1, min_weight_fraction_leaf=0.0, max_features='sqrt', max_leaf_nodes=None, min_impurity_decrease=0.0, min_impurity_split=None, bootstrap=True, oob_score=False, n_jobs=1, random_state=23, verbose=7, warm_start=False, class_weight=None)
```

The reason I used 'Entropy' apart from 'gini' is because there are to many classification that makes up the data. So in order to find good exploratory analysis I used 'entropy. 
Besides it gives me higher accuracy. 

```python
score4=cross_val_score(clf4, train, target, cv=kfold, n_jobs=2)
```

```python
score4
```

```python
round(np.mean(score4)*100,2)
```
